```python
# Получение столбцов всех Датафреймов
df_item.columns
df_item_aliase.columns
df_page.columns
df_property_aliase.columns
df_property.columns
df_statements.columns
```

```python
# Извлечем количесвто псевдонимов для каждого элемента
df_item_with_aliase = df_item.join(df_item_aliase, 'item_id', 'left')
df_count_aliase = df_item_with_aliase.groupBy('item_id').count().sort(desc('count'))
df_item_with_aliase = df_item.join(df_count_aliase, 'item_id', 'left')
df_item_with_aliase = df_item_with_aliase.withColumnRenamed('count', 'count_item_aliase')
df_item_with_aliase.show(4)
```

```python
# Подсчет длинны описания элемента
df_item_with_aliase = df_item_with_aliase.withColumn("en_description_length", functions.length("en_description"))
df_item_with_aliase = df_item_with_aliase.drop(df_item_with_aliase.en_description)
df_item_with_aliase.sort('count_item_aliase')
df_item_with_aliase.show(4)
```

```python
# Извлечем количесвто псевдонимов для каждого свойства
df_property_with_aliase = df_property.join(df_property_aliase, 'property_id', 'left')
df_count_aliase_p = df_property_with_aliase.groupBy('property_id').count().sort(desc('count'))
df_property_with_aliase = df_property.join(df_count_aliase_p, 'property_id', 'left')
df_property_with_aliase = df_property_with_aliase.withColumnRenamed('count', 'count_aliase')
df_property_with_aliase.show(4)
```

```python
# Подсчет длинны описания элемента
df_property_with_aliase = df_property_with_aliase.withColumn("en_description_property_length", functions.length("en_description"))
df_property_with_aliase = df_property_with_aliase.drop(df_property_with_aliase.en_description)
df_property_with_aliase.sort('count_aliase').show(4)
```

```python
# Подсчет количества элементов и свойств на странице
from pyspark.sql.functions import sum,avg,max,count
df_page.groupBy('page_id').count().sort(desc('count')).show(4)

df_page_with_item = df_page.join(df_item_with_aliase, 'item_id', 'left')

df_page_with_item.show(4)
```

```python
# Подсчет количества входящих ссылок к item
df_target_item = df_statements.groupBy('target_item_id').agg(count('source_item_id'))
df_target_item = df_target_item.withColumnRenamed('count(source_item_id)', 'count_in_links_to_item')

df_target_item = df_target_item.join(df_page_with_item, df_target_item.target_item_id == df_page_with_item.item_id, 'left')

df_target_item.show(4)
```

```python
# Подсчет количества выходящих ссылок от item
df_source_item = df_statements.groupBy('source_item_id').agg(count('target_item_id'))
df_source_item = df_source_item.withColumnRenamed('count(target_item_id)', 'count_out_links_to_item')

df_target_item = df_target_item.join(df_source_item, df_source_item.source_item_id == df_target_item.item_id, 'left')

df_target_item = df_target_item.drop('target_item_id')
df_target_item = df_target_item.drop('source_item_id')

df_target_item.show(4)
```

```python
# Удаление столбцов, которые не будут использоваться в дальнейшем
df_target_item = df_target_item.drop('item_id')
df_target_item = df_target_item.drop('page_id')
df_target_item = df_target_item.drop('title')
df_target_item = df_target_item.drop('en_label')
```

```python
df_target_item = df_target_item.withColumn("views",df_target_item.views.cast('int'))
```

# **Обнаружение пропущенных значений и их устранение**

```python
from pyspark.sql.functions import col, count, when

# Создаем список с количеством пустых значений в каждой колонке
empty_counts = [count(when(col(c).isNull() | (col(c) == ""), c)).alias("Empty_" + c) for c in df_target_item.columns]

# Создаем DataFrame с подсчетом пустых значений
missing_data = df_target_item.agg(*empty_counts)

# Преобразуем результат в список строк
result_rows = missing_data.first().asDict()

# Выводим результат в более читаемом формате
for column, count in result_rows.items():
    print(f"{column}: {count}")
```

```python
# Добавления бинарного признака для каждого существующего признака, обозначающего наличие значения в соответствующем признаке
columns_with_missing_data = ['views', 'count_item_aliase', 'en_description_length', 'count_out_links_to_item']
for column in columns_with_missing_data:
    df_target_item = df_target_item.withColumn(column + "_was_is_missing", (when(col(column).isNull() | (col(column) == ""), 1).otherwise(0)) )
df_target_item.select('views', 'views_was_is_missing', 'count_item_aliase', 'count_item_aliase_was_is_missing').show(4)
```

```python
from pyspark.ml.feature import Imputer

# Initialize the Imputer
imputer = Imputer(
    inputCols= columns_with_missing_data, #specifying the input column names
    outputCols=columns_with_missing_data, #specifying the output column names
    strategy="mean"                  # or "median" if you want to use the median value
)
model = imputer.fit(df_target_item)

#Transform the dataset
imputed_df = model.transform(df_target_item)

imputed_df.select(columns_with_missing_data).show(4)
```

```python
from pyspark.sql.functions import col, count, when

# Создаем список с количеством пустых значений в каждой колонке
empty_counts = [count(when(col(c).isNull() | (col(c) == ""), c)).alias("Empty_" + c) for c in imputed_df.columns]

# Создаем DataFrame с подсчетом пустых значений
missing_data = imputed_df.agg(*empty_counts)

# Преобразуем результат в список строк
result_rows = missing_data.first().asDict()

# Выводим результат в более читаемом формате
for column, count in result_rows.items():
    print(f"{column}: {count}")
```

# **Обнаружение выбросов и их устранением**

```python
# Обнаружение выбросов методом межквартильного диапазона (IQR)
clean_df = imputed_df
columns_under_study = ['count_in_links_to_item', 'views', 'count_item_aliase', 'en_description_length', 'count_out_links_to_item']
factor = 1.5
outliers_stats = {}
for column in columns_under_study:
    # Calculate Q1, Q3, and IQR
    quantiles = clean_df.approxQuantile(column, [0.5, 0.95], 0.0)
    q1, q3 = quantiles[0], quantiles[1]
    iqr = q3 - q1

    # Define the upper and lower bounds for outliers
    lower_bound = q1 - factor * iqr
    upper_bound = q3 + factor * iqr

    outliers_count = imputed_df.filter((col(column) < lower_bound) | (col(column) > upper_bound)).count()
    outliers_stats[column] = outliers_count

    # Filter outliers and update the DataFrame
    clean_df = clean_df.filter((col(column) >= lower_bound) & (col(column) <= upper_bound))

print("Количество выбросов: ", outliers_stats)
clean_df.select(columns_under_study).show(4)
```

# **Визуализация распределения наиболее важных признаков**

```python
from pyspark.ml.feature import VectorAssembler, Bucketizer
from pyspark.ml.stat import Correlation

num_buckets = 20
step = 350 / num_buckets
splits = [float("-inf")] + [i * step for i in range(1, num_buckets)] + [float("inf")]
# Построение гистограммы для каждой колонки
for selected_column in columns_under_study:
    bucketizer = Bucketizer(splits=splits, inputCol=selected_column, outputCol="bucketizer")
    df_bucket = bucketizer.transform(imputed_df)
    bucket_counts = df_bucket.groupBy("bucketizer").count().orderBy("bucketizer")
    bucket_counts_pd = bucket_counts.toPandas()
    plt.bar(bucket_counts_pd["bucketizer"], bucket_counts_pd["count"], align="center", label=selected_column)
    plt.title(f"Distribution of {selected_column}")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Бинаризация и логарифмирование данных для визуализации
from pyspark.ml.feature import Binarizer

binarized_columns = {'count_in_links_to_item': 2, 'count_item_aliase': 3}

binarized_df = imputed_df

for column, threshold in binarized_columns.items():
    binarized_df = binarized_df.withColumn(column, binarized_df[column].cast(DoubleType()))
    binarizer = Binarizer(threshold=threshold, inputCol=column, outputCol=column + "_b")
    binarized_df = binarizer.transform(binarized_df)

binarized_df.select('count_in_links_to_item', 'count_item_aliase', 'count_in_links_to_item_b', 'count_item_aliase_b').show(4)
```

# **Корреляций между признакамив**

```python
# Создаем VectorAssembler для объединения признаков в один вектор
vector_assembler = VectorAssembler(inputCols=columns_under_study, outputCol="features")

# Применяем VectorAssembler к данным
data = vector_assembler.transform(imputed_df)

# Вычисляем матрицу корреляции
correlation_matrix = Correlation.corr(data, "features").head()
correlation_array = correlation_matrix[0].toArray()

# Выводим матрицу корреляции
correlation_df = pd.DataFrame(correlation_array, columns=columns_under_study)
```
